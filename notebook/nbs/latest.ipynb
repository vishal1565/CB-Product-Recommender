```python
from tensorflow import keras

import pandas as pd
from pandas import Series, DataFrame
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier

from keras.utils import np_utils
```

```python
banking_product=pd.read_csv('prod_outp.csv')
data=pd.read_csv('complete_dataset_with_indices1.csv', encoding = 'UTF-8')
banking_product = banking_product.rename(columns={'Country' : 'TargetCountry'})
result = pd.merge(banking_product, data , on = ['TargetCountry','Year'])
result.head()
```

```python
result.columns
```

```python
X = result[['Year','TargetCountry','SourceCountry', 'ImportValue', 'Product',
       'weatherIndex', 'geoPoliticalIndex']][:]
y = result[['bills of exchange', 'cash', 'certificates',
       'commitments', 'debt securities', 'derivatives', 'equities securities',
       'finance lease', 'financial guarantees', 'fx derivaties',
       'interest rate derivatives', 'lending and borrowing',
       'letter of credits', 'loans', 'performance guarantees', 'repo',
       'trade financing']][:]
```

```python
from sklearn.preprocessing import LabelEncoder

enc = LabelEncoder()

source_df = enc.fit_transform(X['SourceCountry'])
target_df = enc.fit_transform(X['TargetCountry'])
product_df = enc.fit_transform(X['Product'])

X.drop(['TargetCountry', 'SourceCountry', 'Product'], axis=1)
X['SourceCountry'] = source_df
X['TargetCountry'] = target_df
X['Product'] = product_df
X=X.fillna(0)
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=104, test_size=0.20, shuffle=False)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.multioutput import MultiOutputClassifier
from sklearn.linear_model import LogisticRegression

clf=DecisionTreeRegressor(max_depth=3)
clf.fit(X_train,y_train)
my_array = clf.predict(X_test)
clf.score(X_test,y_test)*100
```
